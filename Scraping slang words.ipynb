```python
import requests
import lxml.html as lh
import pandas as pd
import re
```

```python
url = "https://www.webopedia.com/quick_ref/textmessageabbreviations.asp"

#Create a handle, page, to handle the contents of the website
page = requests.get(url)

#Store the contents of the website under doc
doc = lh.fromstring(page.content)

# table contents were stored under <tbody> tag
#Parse data that are stored between <tbody>..</tbody> of HTML
tr_elements = doc.xpath('//tbody')
```

```python
[len(T) for T in tr_elements[:12]]
```

```python
col=[]
i=0

for t in tr_elements[0]:
    i+=1
    name=t.text_content()
    #print(i,name)
    col.append((name))
```

```python
col
```

#### We can see here that there are \n characters at few places
We need to remove these characters first
To do this we substitue the newline characters with spaces

```python
col_1 = []
```

```python
for i in col:
    col_1.append(re.sub(r"\n"," ",i))
```

```python
for i in range(len(col_1)):
    print(i,col_1[i])
```

#### There were extra spaces at each end of a row 
so we decided to strip those spaces

```python
a = []
for row in col_1:
    a.append(row.strip(" "))
```

```python
a
```

#### As you can see there is space between the slangs and their meaning 
We can split them both using split and create two different lists out of those which can then be used to create our own dataframe

```python
slangs = []
full_forms = []
for row in a:
    row =row.split("  ")
    slangs.append(row[0])
    full_forms.append(row[1])
```

```python
#Creating Slangs Dataframe
slangs_df = pd.DataFrame(slangs,columns=["Slangs"])

slangs_df['Full_Forms'] = full_forms
```
